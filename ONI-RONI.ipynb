```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
roni_url = "https://www.cpc.ncep.noaa.gov/data/indices/RONI.ascii.txt"
roni_df = pd.read_fwf(roni_url)
roni_df.index = pd.date_range(start='1950-01-01', periods=len(roni_df), freq='MS')
print(roni_df.tail())

oni_url = "https://www.cpc.ncep.noaa.gov/data/indices/oni.ascii.txt"
oni_df = pd.read_fwf(oni_url)
oni_df.index = pd.date_range(start='1950-01-01', periods=len(oni_df), freq='MS')
print(oni_df.tail())
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 5))

ax = axs[0]
oni_df['ANOM'].plot(ax=ax, label='ONI');
roni_df['ANOM'].plot(ax=ax, label='RONI');
ax.legend()

ax = axs[1]
ax.bar(roni_df.index, roni_df['ANOM'] - oni_df['ANOM'], width=30, label='RONI - ONI')
ax.legend()
plt.tight_layout()
plt.show()
plt.savefig('.png')
```
